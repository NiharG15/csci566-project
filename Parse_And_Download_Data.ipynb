```python
amusement_excitement_dfs = list()
for csv in DATASET_PATH.glob('amusement*'):
    print('File: ', csv)
    df = pd.read_csv(csv, names=['emotion', 'url', 'disagree', 'agree'])
    display(df.head(n=1))
    amusement_excitement_dfs.append(df)

for csv in DATASET_PATH.glob('excitement*'):
    print('File: ', csv)
    df = pd.read_csv(csv, names=['emotion', 'url', 'disagree', 'agree'])
    display(df.head(n=1))
    amusement_excitement_dfs.append(df)
```

```python
ae_df = pd.concat(amusement_excitement_dfs)
```

```python
ae_df = ae_df.drop_duplicates(subset='url')
```

```python
fear_sadness_dfs = list()
for csv in DATASET_PATH.glob('fear*'):
    print('File: ', csv)
    df = pd.read_csv(csv, names=['emotion', 'url', 'disagree', 'agree'])
    display(df.head(n=1))
    fear_sadness_dfs.append(df)

for csv in DATASET_PATH.glob('sadness*'):
    print('File: ', csv)
    df = pd.read_csv(csv, names=['emotion', 'url', 'disagree', 'agree'])
    display(df.head(n=1))
    fear_sadness_dfs.append(df)
```

```python
fs_df = pd.concat(fear_sadness_dfs)
```
